```python
import requests
import pandas as pd
from time import sleep  # para no saturar la API

# URL base
url = "https://apidatos.ree.es/es/datos/generacion/estructura-generacion"

# Lista para guardar todas las filas
data_total = []

# Rango de a√±os
for year in range(2011, 2023):  # hasta 2022 incluido
    print(f"üìÖ Descargando datos del a√±o {year}...")

    params = {
        "start_date": f"{year}-01-01T00:00:00+01:00",
        "end_date": f"{year}-12-31T23:59:59+01:00",
        "time_trunc": "month"
    }

    response = requests.get(url, params=params)

    # Comprobamos si la API responde bien
    if response.status_code == 200:
        json_data = response.json()

        # A veces puede no haber datos o estar vac√≠o
        if "included" in json_data and json_data["included"]:
            for tech in json_data["included"]:
                nombre = tech["attributes"]["title"]
                valores = tech["attributes"]["values"]

                for v in valores:
                    fila = {
                        "a√±o": year,
                        "tecnologia": nombre,
                        "fecha": v["datetime"],
                        "valor_MWh": v["value"],
                        "porcentaje": v["percentage"]
                    }
                    data_total.append(fila)
        else:
            print(f"‚ö†Ô∏è No hay datos para el a√±o {year}")

    else:
        print(f"‚ùå Error {response.status_code} en el a√±o {year}")

    # Peque√±a pausa para no saturar el servidor
    sleep(0.5)

# Convertimos todo a DataFrame
df = pd.DataFrame(data_total)

# Mostramos el resultado
print(f"\n‚úÖ Datos recopilados: {df.shape[0]} filas")
print(df.head())
```

```python
import requests
from time import sleep
from datetime import datetime, timedelta

diccionario = {"tipo_energia": [], "tipo_generacion": [], "fecha": [], "valor": [], "a√±o": []}

url_base = "https://apidatos.ree.es/es/datos/generacion/estructura-generacion"

for a√±o in range(2011, 2023):
    for mes in range(1, 13):
        # Fecha de inicio y fin del mes
        inicio = datetime(a√±o, mes, 1)
        if mes == 12:
            fin = datetime(a√±o + 1, 1, 1) - timedelta(seconds=1)
        else:
            fin = datetime(a√±o, mes + 1, 1) - timedelta(seconds=1)

        start_date = inicio.strftime("%Y-%m-%dT00:00:00+01:00")
        end_date = fin.strftime("%Y-%m-%dT23:59:59+01:00")

        params = {
            "start_date": start_date,
            "end_date": end_date,
            "time_trunc": "month"
        }

        response = requests.get(url_base, params=params)

        if response.status_code == 200:
            json_data = response.json()

            if "included" not in json_data or not json_data["included"]:
                print(f"‚ö†Ô∏è Sin datos {a√±o}-{mes:02d}")
                continue

            for elemento in json_data["included"]:
                for dato in elemento["attributes"]["values"]:
                    diccionario["tipo_energia"].append(elemento["attributes"]["title"])
                    diccionario["tipo_generacion"].append(elemento["attributes"].get("type", "Desconocido"))
                    diccionario["valor"].append(dato["value"])
                    diccionario["fecha"].append(dato["datetime"])
                    diccionario["a√±o"].append(a√±o)

            print(f"‚úÖ {a√±o}-{mes:02d} descargado correctamente")

        else:
            print(f"‚ùå Error {response.status_code} en {a√±o}-{mes:02d}")

        # Peque√±a pausa para no saturar la API
        sleep(0.3)
```

```python
import pandas as pd
df = pd.DataFrame(diccionario)
print(df.shape)
print(df.head())
```

```python
import requests
import pandas as pd
from time import sleep

# Diccionario de comunidades
cod_comunidades = {
    "Andaluc√≠a": 8744,
    "Arag√≥n": 8745,
    "Asturias": 8746,
    "Baleares": 8747,
    "Canarias": 8748,
    "Cantabria": 8749,
    "Castilla-La Mancha": 8750,
    "Castilla y Le√≥n": 8751,
    "Catalu√±a": 8752,
    "Comunidad Valenciana": 8753,
    "Extremadura": 8754,
    "Galicia": 8755,
    "Madrid": 8756,
    "Murcia": 8757,
    "Navarra": 8758,
    "Pa√≠s Vasco": 8759,
    "La Rioja": 8760
}

# Estructura del diccionario de resultados
diccionario_ccaa = {
    "tipo_energia": [],
    "tipo_generacion": [],
    "fecha": [],
    "valor": [],
    "ccaa": []
}

url_base = "https://apidatos.ree.es/es/datos/generacion/estructura-generacion"

# üîÅ Bucle por comunidad
for nombre, codigo in cod_comunidades.items():
    print(f"üìç Descargando datos de {nombre}...")

    params = {
        "start_date": "2022-01-01T00:00:00+01:00",
        "end_date": "2022-12-31T23:59:59+01:00",
        "time_trunc": "month",
        "geo_trunc": "ccaa",
        "geo_ids": codigo
    }

    response = requests.get(url_base, params=params)

    if response.status_code == 200:
        data = response.json()

        if "included" in data:
            for elemento in data["included"]:
                tipo_energia = elemento["attributes"]["title"]
                tipo_generacion = elemento["attributes"].get("type", "Desconocido")

                for valor in elemento["attributes"]["values"]:
                    diccionario_ccaa["tipo_energia"].append(tipo_energia)
                    diccionario_ccaa["tipo_generacion"].append(tipo_generacion)
                    diccionario_ccaa["valor"].append(valor["value"])
                    diccionario_ccaa["fecha"].append(valor["datetime"])
                    diccionario_ccaa["ccaa"].append(nombre)
        else:
            print(f"‚ö†Ô∏è Sin datos para {nombre}")
    else:
        print(f"‚ùå Error {response.status_code} en {nombre}")

    sleep(0.5)  # evitar sobrecargar la API

# Convertimos el diccionario en DataFrame
df_ccaa = pd.DataFrame(diccionario_ccaa)
print(f"\n‚úÖ Datos descargados: {df_ccaa.shape[0]} filas")
print(df_ccaa.head())
```
